```python
def get_all_files(root_dir_list, ext_list):
    if not isinstance(ext_list, list):
        ext_list = [ext_list]
    if not isinstance(root_dir_list, list):
        ext_list = [root_dir_list]
        
    all_files = []
    for root_dir in root_dir_list:
        for root, dirs, files in os.walk(root_dir, topdown=False):
            for file in files:
                for target_ext in ext_list:
                    ext_len = len(target_ext)
                    ext = file[-ext_len:]
                    if ext == target_ext:
                        all_files.append(os.path.join(root, file))
    return all_files
```

```python
core_place_dir = [r'/media/gpu/easystore/Safegraph/Core Places US (Nov 2020 - Present)',
                  r'/media/gpu/easystore/Safegraph/Core Places US (Pre-Nov-2020)']


all_core_place_files = get_all_files(root_dir_list=core_place_dir, ext_list=['.gz'])

all_core_place_files = natsorted(all_core_place_files, reverse=True)

all_core_place_files = all_core_place_files[40:] + all_core_place_files[:40]
all_core_place_files
```

# merge all historical POIs

```python
def get_core_place_month(file_name):
    f = file_name
    tag = 'Pre-Nov-2020'
    if tag in f:
        tag_pos = f.find(tag)
        month = f[tag_pos + len(tag) + 2:tag_pos + len(tag) + 6] + "-" + f[tag_pos + len(tag) + 7:tag_pos + len(tag) + 9]
        
    tag = '(Nov 2020 - Present)'
    if tag in f:
        tag_pos = f.find(tag)
        month = f[tag_pos + len(tag) + 10:tag_pos + len(tag) + 14] + "-" + f[tag_pos + len(tag) + 15:tag_pos + len(tag) + 17]        
       
    return month

get_core_place_month(file_name=all_core_place_files[0])
```

```python
df_all = pd.read_csv(all_core_place_files[0])
df_all['month'] = get_core_place_month(all_core_place_files[0])

for idx, f in enumerate(all_core_place_files[1:]):    
    month = get_core_place_month(f)
    df = pd.read_csv(f)
    print(f'Processing {len(df)} rows from file # {idx} / {len(all_core_place_files)}: {f}')
    
    df['month'] = month
    old_row_cnt = len(df_all)
    df_all = pd.concat([df_all, df]).drop_duplicates(subset='placekey').reset_index(drop=True)
    new_row_cnt = len(df_all)
    print(f'Current rows counts: {new_row_cnt}, previous rows: {old_row_cnt}. Added {new_row_cnt - old_row_cnt}, removed {old_row_cnt + len(df) - new_row_cnt} duplated rows.')
    #df = df.set_index('placekey')
    print()

df_all.to_csv(r'/media/gpu/easystore/Safegraph/Core Places US (Nov 2020 - Present)/core_poi/POI_merged_202106.csv', index=False)
    
```
