```python
def extract_neighorhood_pattern(df, fips):
    column_name = "device_home_areas"
    idxs  = df.apply(is_fips_in_row, args=(column_name, fips), axis=1)
    return df[idxs]
    
def is_fips_in_row(row, column_name, fips):
    return is_fips_in_dict(row[column_name], fips)
    
def is_fips_in_dict(poi_dict, fips):
    fips = str(fips)
    fips_len = len(fips)
    for key in poi_dict.keys():
        key = str(key)
        if key[:fips_len] == fips:
            return True
    return False
```

```python
import subprocess
import os

from tqdm import tqdm

def get_all_files(root_dir_list, ext_list):
    if not isinstance(ext_list, list):
        ext_list = [ext_list]
    if not isinstance(root_dir_list, list):
        ext_list = [root_dir_list]
        
    all_files = []
    for root_dir in root_dir_list:
        for root, dirs, files in os.walk(root_dir, topdown=False):
            for file in files:
               #  print(file)
                for target_ext in ext_list:
                    ext_len = len(target_ext)
                    ext = file[-ext_len:]
                    if ext == target_ext:
                        all_files.append(os.path.join(root, file))
    return all_files
```

```python
import subprocess
import os

from tqdm import tqdm
import multiprocessing as mp

import glob
import os
import pandas as pd


def groupy_counties(edge_csv_list):
    if not isinstance(edge_csv_list, list):
        edge_csv_list= [edge_csv_list]
        
    while len(edge_csv_list) > 0:
        edge_csv_file = edge_csv_list.pop(0)
        print("Processing:", edge_csv_file)
        df = pd.read_csv(edge_csv_file, dtype={'visitor_home_cbgs': str})
        saved_path = r'H:\Safegraph_reorganized\county_weekly_patterns'

        df['county_code'] = df['visitor_home_cbgs'].str.zfill(12).str[:5]
        county_list = df['county_code'].unique()


        for idx, county in tqdm(enumerate(county_list)):
            basename = os.path.basename(edge_csv_file)
            new_name = f'County_{county}_{basename}'
            new_name = os.path.join(saved_path, new_name)
            county_df = df[df['county_code'] == county]
            county_df.to_csv(new_name, index=False)

     

def patterns_CBG_all_csv(file_dir = r'H:\Safegraph', process_cnt = 8):  
    
    out = subprocess.getoutput(f"dir  /b/n/s {file_dir}\weekly*edges*.csv")

    all_files = out.split("\n")
    
    if process_cnt == 1:
         groupy_counties(all_files)
    
    if process_cnt > 1:        
        all_files_mp = mp.Manager().list()
        for f in all_files:
            all_files_mp.append(f)
       
    pool = mp.Pool(processes=process_cnt)
    for i in range(process_cnt):
        pool.apply_async(groupy_counties, args=(all_files_mp,))
    pool.close()
    pool.join()

patterns_CBG_all_csv(file_dir = r'H:\Safegraph')        
```
